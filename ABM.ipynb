# Agent-based modelling

```python
import pandas as pd
from ABM_functions import *
```

```python
import warnings

# To ignore all warnings
warnings.filterwarnings("ignore")

# setting random seed
np.random.seed(42)
```

#### Setting up parameters

```python
attachment = ['Normal Job', 'Flex Job', 'Sick Leave', 'Job Clarification', 'Disability Pension', 'Retired'] # types of attachment to labour market
n = [1000, 1000, 1000, 1000, 1000, 1000] # all people should start out in 'Normal job'. But in order for them to move around, all attachment groups are assigned 1000 spaces 

weights = {# based on pcss-scale symptoms, from scale 0-6
    'symptoms': {
        'headache',
        'nausea',
        'dizziness',
        'fatigue',
        'sleep problems',
        'light sensitivity',
        'noise sensitivity',
        'irritability',
        'sadness',
        'feeling mentally foggy',
        'difficulty concentrating',
        'difficulty remembering',
        'visual problems'
    }
}
```

## Running the agent-based model: Baseline

```python
for y in range(10):
    run_abm(weeks = 52*5,
            attachment = attachment,
            n = n, 
            weights = weights,
            plot_each_tick = False,
            intervention = None,
            percentage = True,
            run_id = y
            )
```

### Running the ABM: Intervention

```python
for y in range(10):
    run_abm(weeks = 52*5,
            attachment = attachment, # the job titles in the company
            n = n, # the number of employees with each of the titles
            weights = weights, # the weights for generating agents
            plot_each_tick = False,
            intervention = 'intervention',
            percentage = True,
            run_id = y
            )
```
